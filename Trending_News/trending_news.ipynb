```python
from GoogleNews import GoogleNews
```

```python
googlenews=GoogleNews()
# choose language 
googlenews = GoogleNews(lang='en')
#choose period
googlenews=GoogleNews(period='7d')
#choose encoding
googlenews = GoogleNews(encode='utf-8')
# news by location
googlenews = GoogleNews(region='India')
result=googlenews.result()
result
```

```python
for i in result:
     print("Title: ",i['title'])
     print("Date / Time: ",i['date'])
     print("Description: ",i['desc'])
     print("Link: ",i['link'])
```

```python
# custom period mm/dd/yyyy
googlenews=GoogleNews(start='12/05/2021',end='12/07/2021')
# search by news.google.com
googlenews.get_news('TATA')
result=googlenews.result()
result
```

```python
#google.com section news search 
googlenews.set_lang('en')
googlenews.set_time_range('12/06/2020','12/07/2020')
googlenews.set_encode('utf-8')
googlenews.search('Maharashtra')
# prints news count
print(googlenews.total_count())
result=googlenews.result()
result
```
